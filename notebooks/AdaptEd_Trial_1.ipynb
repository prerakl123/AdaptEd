<h4>Literature Survey</h4>

<p>
<strong>Personalized Learning: A Growing Need </strong>

The traditional, one-size-fits-all approach to education struggles to cater to the diverse needs and learning styles of students in modern classrooms<sup><a href="">[1]</a></sup>.  Research emphasizes the importance of personalized learning, which tailors instruction and materials to individual strengths and weaknesses<sup><a href="">[2]</a></sup>. This approach fosters deeper engagement, improves knowledge retention, and promotes a more effective learning experience<sup><a href="">[3]</a></sup>.
</p>

<p>
<strong>The Rise of Generative AI in Education</strong>

Generative AI, a branch of artificial intelligence focused on creating new content, holds immense potential for personalized learning<sup><a href="">[4]</a></sup>. These models can analyze vast amounts of educational data and student performance to generate customized learning materials, including exercises, explanations, and adaptive learning pathways<sup><a href="">[5]</a></sup>.
</p>

<p>
<strong>Existing Work on Generative AI for Personalized Learning</strong>

Several studies have explored the use of Generative AI in educational settings. For example, research by Passi et al. (2021) demonstrates the effectiveness of AI-powered tutors that personalize explanations based on student understanding<sup><a href="">[6]</a></sup>. Additionally, Tian et al. (2023) explore the application of Generative AI to create customized learning materials in specific domains like language learning <sup><a href="">[7]</a></sup>.
</p>
